```python

class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        try:
            if amount<=0:
                raise ValueError("Amount cannot be less than or equal to 0.")
            if amount > self.balance:
                raise ValueError("Insufficient funds! Withdrawal amount exceeds balance.")
            self.balance -= amount
            return f"Withdrawal successful! Remaining balance: {self.balance}"
        except ValueError as e:
            return f"Error: {e}"

account = BankAccount(500)

print(account.withdraw(-10))
print(account.withdraw(300))
```
