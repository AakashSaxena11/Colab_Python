```python
data = {'Cab_ID': [1,2,3], 'Driver': ['Alice', 'Bob', 'Charlie'], 'Fare': [15.5, 20.0, 30.0]}
df = pd.DataFrame(data)
print("Original Dataframe:\n",df)
```

```python
print(df.loc[1])
```

```python
print(df.iloc[2])
```

```python
df['Discounted_Fare'] = df['Fare'] * 0.9
print("\nDataFrame after operation:\n",df)
mean_fare = df['Fare'].mean()
print("\nMean Fare:\n",mean_fare)
```

```python
df.loc[1, 'Fare'] = None
print("\nDataFrame after modification:\n",df)

print("\nMissing Data Check:\n", df.isna())

df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
print("\nDataFrame after filling missing values:\n",df)
```

```python
import matplotlib.pyplot as plt
car = ["A", "B", "C", "D", "E", "F"]
weight = [0.48, 1.7, 2, 2, 2.3, 3]

data = {'Car': car, 'Weight': weight}
df = pd.DataFrame(data)

df.plot(x='Car', y='Weight', kind = 'bar', color = 'blue')
plt.xlabel('Car')
plt.ylabel('Weight')
plt.title('Weight of Cars')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
car = ["A", "B", "C", "D", "E", "F"]
weight = [0.48, 1.7, 2, 2, 2.3, 3]

data = {'Car': car, 'Weight': weight}
df = pd.DataFrame(data)

df.plot(x='Car', y='Weight', kind = 'line', color = 'green', marker = "*")
plt.xlabel('Car')
plt.ylabel('Weight')
plt.title('Weight of Cars')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
car = ["A", "B", "C", "D", "E", "F"]
weight = [0.48, 1.7, 2, 2, 2.3, 3]

data = {'Car': car, 'Weight': weight}
df = pd.DataFrame(data)

df.plot(x='Car', y='Weight', kind = 'scatter', color = 'red')
plt.xlabel('Car')
plt.ylabel('Weight')
plt.title('Weight of Cars')
plt.grid(True)
plt.tight_layout()
plt.show()
```
