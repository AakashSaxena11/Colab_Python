```python
def FileNot(FN):
    try:
        with open(FN, "r") as file:
            content = file.read()
            print(content)
    except FileNotFoundError:
        print("File not found")
```

```python
import subprocess

try:
    process = subprocess.Popen(['echo', 'H'], stdout=subprocess.PIPE)
    output, _ = process.communicate()
    print("Output from pipe:")
    print(output.decode('utf-8'))
except Exception as e:
    print(f"Error occurred using pipe: {e}")
```
