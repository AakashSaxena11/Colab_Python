```python
class Vehicle:
    def __init__(self, make, model, year, color):
        self.make = make
        self.model = model
        self.year = year
        self.color = color

    def description(self):
        return f"{self.year} {self.color} {self.make} {self.model}"

class Car(Vehicle):
    def __init__(self, make, model, year, color, num_doors):
        super().__init__(make, model, year, color)
        self.num_doors = num_doors

    def description(self):
        return f"{super().description()} with {self.num_doors} doors"

    def honk(self):
        print("Honk!")

class Motorcycle(Vehicle):
    def __init__(self, make, model, year, color, engine_type):
        super().__init__(make, model, year, color)
        self.engine_type = engine_type

    def description(self):
        return f"{super().description()} with a {self.engine_type} engine"

    def rev_engine(self):
        print("Vroom!")

my_car = Car("XUV 700", "AX7L", 2022, "White", 4)
my_motorcycle = Motorcycle("Harley-Davidson", "Sportster", 2022, "Black", "V-twin")

print(my_car.description())
my_car.honk()

print(my_motorcycle.description())
my_motorcycle.rev_engine()
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"{self.name} is {self.age} years old.")

class PG(Student):
    def __init__(self, name, age, college):
        super().__init__(name, age)
        self.college = college

    def display_info(self):
        super().display_info()
        print(f"He is studying in {self.college}")

ob = PG("Aakash", 22, "VIT")
ob.display_info()
```

```python
class Book:
  def __init__(self, title, author, pages):
    self.title = title
    self.author = author
    self.pages = pages

  def __str__(self):
    return f"{self.title} by {self.author}"

  def __repr__(self):
    return f"Book(title='{self.title}', author='{self.author}', pages={self.pages})"

  def __len__(self):
    return self.pages

  def __eq__(self, other):
    if not isinstance(other, Book):
      return NotImplemented
    return (self.title == other.title) and (self.author == other.author) and (self.pages == other.pages)

  def __lt__(self, other):
    if not isinstance(other, Book):
      return NotImplemented
    return self.pages < other.pages

b1 = Book("The Alchemist", "Paulo Coelho", 150)
b2 = Book("Meaning of Life", "Franklin", 200)
print(b1)
print(str(b2))
print(repr(b1))
print(len(b1))
print(b1 == b2)
print(b1 < b2)
```
